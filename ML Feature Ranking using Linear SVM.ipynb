```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn import svm
import pandas as pd
```

```python
train = pd.read_csv("reged0_text/reged0_train.data",delimiter=" ")
label = pd.read_csv("reged0_text/reged0_train.targets",delimiter=" ")
```

```python
X = train.iloc[:,:-1]
print X.shape
Y = label.iloc[:]
print Y.shape
```

```python
clf = svm.SVC(kernel='linear', C = 1.0)
```

```python
clf.fit(X,Y)
```

```python
w = clf.coef_[0]
print(w)
```

```python
C = []
for i in range(len(w)):
    C.append(pow(w[i],2))
```

```python
print C
```

```python
len(C)
```

```python
wght = {}
for i in range(len(C)):
    wght[i] = C[i]
```

```python
print wght
```

```python
import operator
sorted_w = sorted(wght.items(),key = operator.itemgetter(1))
```
